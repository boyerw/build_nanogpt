```python
from transformers import GPT2LMHeadModel
import matplotlib.pyplot as plt
```

```python
# !pip install transformers -U
```

```python
model_hf = GPT2LMHeadModel.from_pretrained("gpt2") # 124M
sd_hf = model_hf.state_dict()

for k, v, in sd_hf.items():
    print(k, v.shape)
```

# Visualizing positional encodings

```python
%matplotlib inline

plt.imshow(sd_hf["transformer.wpe.weight"], cmap='RdYlBu', vmin=-0.08, vmax=0.08)
plt.colorbar(orientation='vertical')
plt.show()
```

```python
%matplotlib inline

fig, axes = plt.subplots(3, 2, figsize=(10, 10))
axes = axes.flatten()

axes[0].plot(sd_hf["transformer.wpe.weight"][:, 50])
axes[1].plot(sd_hf["transformer.wpe.weight"][:, 100])
axes[2].plot(sd_hf["transformer.wpe.weight"][:, 150])
axes[3].plot(sd_hf["transformer.wpe.weight"][:, 200])
axes[4].plot(sd_hf["transformer.wpe.weight"][:, 250])
axes[5].plot(sd_hf["transformer.wpe.weight"][:, 300])

fig.suptitle('Positional encoding curve for each dimension')
plt.tight_layout()
plt.show()
```

```python
%matplotlib inline

fig, axes = plt.subplots(3, 2, figsize=(10, 10))
axes = axes.flatten()

axes[0].plot(sd_hf["transformer.wpe.weight"][0, :])
axes[1].plot(sd_hf["transformer.wpe.weight"][200, :])
axes[2].plot(sd_hf["transformer.wpe.weight"][400, :])
axes[3].plot(sd_hf["transformer.wpe.weight"][600, :])
axes[4].plot(sd_hf["transformer.wpe.weight"][800, :])
axes[5].plot(sd_hf["transformer.wpe.weight"][1000, :])
```
